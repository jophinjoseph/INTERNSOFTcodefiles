```python
jophin=pd.Series(np.random.normal(size=40))
print(jophin)
jophin.plot()
```

```python
jophin.cumsum().plot(grid=True)
```

```python
vari=pd.DataFrame({'norma':np.random.normal(size=20),
                  'gamma':np.random.gamma(2,size=20),
                  'poisson':np.random.poisson(size=20)
                 })
print(var)
vari.cumsum(0).plot()
```

```python
vari.cumsum(0).plot(subplots=True)
```

```python
vari.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,var in enumerate(['norma','gamma','poisson']):
    vari[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
fig=plt.figure()
ax=plt.axes()
s=np.linspace(0,10,50)
print(s)
ax.plot(s,np.sin(s));
```

```python
plt.plot(s,np.sin(s))
```

```python
#line colors and styles
```

```python
plt.plot(s, np.sin(s - 0), color='green')        
plt.plot(s, np.sin(s - 1), color='b')           
plt.plot(s, np.sin(s - 2), color='0.55')  
```

```python
plt.plot(s, np.cos(s - 0), linestyle='--')        
plt.plot(s, np.sin(s - 1), linestyle=':')           
plt.plot(s, np.sin(s - 2), linestyle='-.')  
```

```python
plt.style.use('seaborn-white')
data=np.random.randn(10000)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins=8, normed=True, alpha=0.8,
         histtype='stepfilled')
```

```python
x = np.random.normal(0, 2, 1000)
y = np.random.normal(-2, 5, 1000)

jop = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **jop)
plt.hist(y, **jop)
```

```python
counts, bin_edges = np.histogram(data, bins=8)
print(counts)
```

```python
#2d histograms
```

```python
mean=[5,2]
cov=[[3,1],[2,0.5]]
x,y=np.random.multivariate_normal(mean,cov,10000)
```

```python
plt.hist2d(x, y, bins=10, cmap='Reds')
cb = plt.colorbar()
```

```python
plt.hexbin(x, y, gridsize=10, cmap='Blues')

cb = plt.colorbar()
```
